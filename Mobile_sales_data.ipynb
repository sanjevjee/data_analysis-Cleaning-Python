## Data Preprocessing

```python
# checking outlier in df
df.boxplot(figsize=(12,12))
plt.show()
```

- there are few outlier in rating and actual price assuming all are real data point.

```python
# checking null values
df.isna().sum()
```

- there are some missing values in Actual price, Ram(GB), Storage (GB), Camera.

```python
# imputing all missing values in Camera, Storage (GB) , RAM (GB),Discount price,actual price

# Impute missing values in 'Camera' with the most frequent value
df['Camera'] = df['Camera'].fillna(df['Camera'].mode()[0])

# Impute missing values in 'Storage (GB)' with the median
df['Storage (GB)'] = df['Storage (GB)'].fillna(df['Storage (GB)'].median())

# Impute missing values in 'RAM (GB)' with the median
df['RAM (GB)'] = df['RAM (GB)'].fillna(df['RAM (GB)'].median())

# Impute missing values in 'Discount price' with the mean
df['Discount price'] = df['Discount price'].fillna(df['Discount price'].mean())

# Impute missing values in 'Actual price' with the mean
df['Actual price'] = df['Actual price'].fillna(df['Actual price'].mean())
```

- missing value has been imputed.
